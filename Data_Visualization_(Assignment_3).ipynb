```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import missingno as msn
```

```python
airbnb = pd.read_csv('/content/Assignment 3 airbnb_dataset.csv')
```

```python
HR_dataset = pd.read_csv("/content/Assignment 3 HR_ dataset.csv")
```

```python
airbnb.isnull()
```

```python
## Dropping the rows having mising values
airbnb.dropna(inplace=True)
```

```python
## Converting the last review column to datetime format
airbnb['last_review'] = pd.to_datetime(airbnb['last_review'])
```

```python
## Droping the unwanted column
airbnb = airbnb.drop(['host_name', 'id'], axis =1)
```

```python
## Saving the cleaned dataset to a new file
airbnb.to_csv('/content/cleaned_dataset.csv', index=False)
```

```python
## Checking the cleaned dataset
airbnb.head(10)
```

```python
## Calculating the no. of days since the last review
airbnb['days_since_last_review'] = (pd.to_datetime('2023-06-10') - airbnb['last_review']).dt.days
```

```python
neighbourhood_group_count = airbnb.groupby('neighbourhood_group').count().head()
neighbourhood_group_count
```

```python
## Creating Histogram chart
plt.figure(figsize=(10,6))
plt.hist(airbnb['price'], bins=30, edgecolor ='black')
plt.xlabel('Price')
plt.ylabel('Count')
plt.title('Distribution of Prices')
plt.show()
```

```python
## Creating Boxplot graph
plt.figure(figsize=(10,6))
sns.boxplot(x='neighbourhood_group', y = 'price', data=airbnb)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Price Distribution across Neighbourhood Groups')
plt.show()
```

```python
## Creating Heat map Chart
plt.figure(figsize=(10,6))
sns.heatmap(airbnb[['neighbourhood_group', 'room_type', 'price']].corr(), annot=True, cmap='viridis')
plt.title('Correaltion Heatmap')
plt.show()
```

```python
counts = airbnb['neighbourhood_group'].value_counts()
plt.figure(figsize=(10,6))
plt.bar(counts.index, counts.values)
plt.xlabel('Neighbourhood Groups')
plt.ylabel('Count')
plt.title('Neighbourhood Group Counts')
plt.show()
```

```python
## Creating Scatter Chart
plt.figure(figsize=(10,5))
sns.barplot(x='room_type', y = 'price', data = airbnb)
plt.xlabel('Room Type')
plt.ylabel('Price')
plt.title('Price by Room Type')
plt.show()
```

```python
## Creating Pie Chart
room_type_counts = airbnb['room_type'].value_counts()
plt.figure(figsize=(8,6))
plt.pie(room_type_counts, labels=room_type_counts.index, autopct = '%1.1f%%')
plt.title('Room Type Proportions')
plt.show()
```

```python
## Creating Violin Graph
plt.figure(figsize=(10,6))
sns.violinplot(x='neighbourhood_group', y = 'price', data=airbnb)
plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')
plt.title('Price istribution by Neighbourhood Group')
plt.show()
```

```python
## Creating  bar plot 
avg_price_by_neighbourhood = airbnb.groupby('neighbourhood')['price'].mean().sort_values(ascending=False)
plt.figure(figsize=(30,8))
plt.bar(avg_price_by_neighbourhood.index, avg_price_by_neighbourhood.values)
plt.xlabel('Neighbourhood')
plt.ylabel('Average Price')
plt.title('Average Price by Neighbourhood')
plt.xticks(rotation=45)
plt.show()
```

```python
## Creating Line chart
airbnb['last_review'] = pd.to_datetime(airbnb['last_review'])
avg_price = airbnb.groupby(airbnb['last_review'])['price'].mean()
plt.plot(avg_price.index, avg_price.values)
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.title('Average Price Over Time')
plt.xticks(rotation=45)
plt.show()
```

```python
## dropping null values
HR_dataset.dropna(inplace=True)
```

```python
## Converting the Date of hire column in datetime format
HR_dataset['DateofHire'] = pd.to_datetime(HR_dataset['DateofHire'])
```

```python
## Creating Bubble Chart
x = HR_dataset['DOB']
y = HR_dataset['Salary']
size = HR_dataset['EmpSatisfaction']* 100
plt.figure(figsize=(30,8))
plt.scatter(x,y, s = size, alpha = 0.5)
plt.xlabel('DOB')
plt.ylabel('Salary')
plt.title('Bubble Chart: Age vs Salary')
plt.colorbar(label='EmpSatisfaction Level')
plt.show()
```

```python
## draw donut chart
Department_counts = HR_dataset['Department'].value_counts()
plt.figure(figsize=(15,6))
plt.pie(Department_counts, labels = Department_counts.index, autopct = '%1.1f%%', startangle = 90)
plt.title('Department Distribution')
circle = plt.Circle((0,0),0.6, color ='white')
plt.gca().add_artist(circle)
plt.show()
```

```python
## Creating Pictograph visalization
Department_counts = HR_dataset['Department'].value_counts()
pictograms = ['⚕️','✈️', '⚙️', '⚖️', '⚽️']
labels = Department_counts.index
fig, ax = plt.subplots(figsize=(10,6))
ax.bar(labels, Department_counts,color='skyblue')
ax.set_xticklabels(pictograms)
ax.set_ylabel('Count')
ax.set_title('Pictograph of Department Distribution')
plt.show()
```
